```python
df2 = df[["date","China","Italy","Brazil","Spain","US","United Kingdom","France","Turkey"]]
```

```python
df.info()
```

```python
df2.info()
```

```python
df2.set_index("date",inplace=True)     #df2 = df2.set_ind.... yapmamak için inplace = True yaptık!
```

```python
df2.info()
```

```python
import bar_chart_race as bcr
```

```python
df_total = df2.cumsum(axis=0)
```
